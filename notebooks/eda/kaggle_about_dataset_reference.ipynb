## 1. Visao geral do dataset

- Nome: **Synthetic Logistics Operations Database (2022-2024)**
- Tema: operacoes logisticas de uma transportadora de carga pesada (Class 8 trucking)
- Horizonte temporal: **janeiro de 2022 a dezembro de 2024**
- Natureza: **dataset sintetico**, construido para parecer realista e util em analytics, SQL, BI, engenharia de dados e data science
- Escopo: base operacional completa, com entidades de referencia, transacoes operacionais e tabelas agregadas
- Relacionamentos: as tabelas possuem **integridade referencial** e foram pensadas para JOINs e analises multi-tabelas

## 2. O que existe dentro do dataset

Segundo a pagina do Kaggle, a base representa uma empresa ficticia de transporte rodoviario com mais de **85 mil registros** distribuidos em **14 tabelas interconectadas**.

O conjunto cobre, entre outros temas:

- motoristas e atribuicoes
- caminhoes e reboques
- clientes e instalacoes
- rotas e cargas
- viagens realizadas
- abastecimentos
- manutencao
- eventos de entrega
- incidentes de seguranca
- metricas agregadas de motoristas e utilizacao de caminhoes

## 3. Publico-alvo

A pagina destaca que o dataset foi pensado para:

- praticantes de SQL que querem trabalhar com CTEs, window functions e JOINs realistas
- analistas de dados interessados em metricas operacionais, custo por milha, scorecards e utilizacao de frota
- profissionais migrando para supply chain analytics
- estudantes de data science que queiram modelar manutencao, retencao de motoristas e otimizacao de rotas
- pessoas em transicao de carreira, principalmente de operacoes para analytics

## 4. Motivacao do dataset

O criador argumenta que bases de logistica costumam cair em dois extremos:

- sao proprietarias e nao podem ser abertas
- ou sao simplificadas demais e pouco realistas

Este dataset busca preencher esse espaco, oferecendo complexidade operacional sem depender de dados confidenciais.

A pagina cita alguns padroes realistas incorporados ao dataset:

- pico de preco do diesel em 2022 e queda em 2023-2024
- turnover de motoristas em torno de 15% ao ano
- utilizacao de equipamentos na faixa de 65%
- pontualidade entre 85% e 95%
- manutencao alinhada a cronogramas tipicos de Class 8

## 11. Limitacoes conhecidas

### Nao incluidos

- Hours of Service (HOS)
- modelagem de clima
- historico de pagamento de clientes
- detalhes de sinistros de seguro
- manifestos detalhados de carga
- dados de ELD

### Simplificados

- incidentes de seguranca foram reduzidos a informacoes basicas
- descricoes de manutencao foram generalizadas
- contratos de clientes foram simplificados
- planejamento de rota nao inclui trafego ou obras

A pagina deixa claro que essas simplificacoes foram intencionais para manter valor analitico sem complexidade excessiva.
